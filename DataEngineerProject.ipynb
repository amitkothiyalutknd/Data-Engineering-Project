```python
# AccCountClean = AccCount[["Code", "Short Name", "Table Name", "Long Name", "Currency Unit", "Region"]]
account_country_table_create = (
                                "CREATE TABLE IF NOT EXISTS `accountCountry` ("
                                 " `countryCode` varchar(10),"
                                 " `shortName` varchar(40),"
                                 " `tableName` varchar(40),"
                                 " `longName` varchar(100),"
                                 " `currencyUnit` varchar(40),"
                                 " `Region` varchar(30),"
                                 " PRIMARY KEY (`countryCode`)"
                                ") ENGINE=InnoDB")
cur.execute(account_country_table_create)
conn.commit()
```

```python
account_data_table_create = (
                                "CREATE TABLE IF NOT EXISTS `accountdata` ("
                                 " `countryName` varchar(40),"
                                 " `countryCode` varchar(10),"
                                 " `indicatorName` varchar(100),"
                                 " `indicatorCode` varchar(100),"
                                 " `year1995` varchar(100),"
                                 " `year2000` varchar(100),"
                                 " `year2005` varchar(100),"
                                 " `year2010` varchar(100),"
                                 " `year2014` varchar(100)"
                                ") ENGINE=InnoDB")
cur.execute(account_data_table_create)
conn.commit()
```

```python
account_series_table_create = (
                                "CREATE TABLE IF NOT EXISTS `accountSeries` ("
                                 " `seriesCode` varchar(40),"
                                 " `topic` varchar(40),"
                                 " `indicatorName` varchar(100),"
                                 " `longDefinition` LONGTEXT"
                                ") ENGINE=InnoDB")
cur.execute(account_series_table_create)
conn.commit()
```
